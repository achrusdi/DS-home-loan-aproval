# **Import Library**

```python
  import pandas as pd
  import numpy as np
  import matplotlib.pyplot as plt
  import seaborn as sns
```

# **Import Dataset**

```python
df = pd.read_csv('loan_sanction_train.csv')
df.head()
```

# **Data Preparation**

## Check Data Types

```python
df.dtypes
```

## Check Missing Values

```python
lst_missval = []

for i in df.isnull().sum():
  lst_missval.append((i / len(df) * 100))

df_missval = pd.DataFrame({'Column Name':df.columns,
                           'Missing Value Percentage (%)':np.round(lst_missval,2),
                           'Data Types':df.dtypes})

df_missval = df_missval[df_missval['Missing Value Percentage (%)'] != 0]

df_missval.sort_values(by='Missing Value Percentage (%)', 
                       ascending = False).reset_index().drop(columns = 'index')
```

```python
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].median())
df['LoanAmount'] = df['LoanAmount'].fillna(df['LoanAmount'].median())
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median())
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
```

## Drop Unrelevant Variables

```python
df.head()
```

```python
df_model = df.drop(columns = ['Loan_ID'])
df_model.head()
```

## Encoding

```python
lst_encode = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed',
              'Property_Area']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in lst_encode:
  df_model[i] = le.fit_transform(df_model[i])

df_model.head()
```

# **Modelling**

## Define X & Y Variables

```python
X = df_model.drop(columns = ['Loan_Status']).values
y = df_model['Loan_Status'].values
```

## Split Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

## Training Model

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)

y_pred_nb = nb_classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_nb).ravel()

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
akurasi_nb = round(accuracy_score(y_test, y_pred_nb),2)
precision_nb = round(precision_score(y_test, y_pred_nb, average='weighted'),2)
recall_nb = round(recall_score(y_test, y_pred_nb, average='weighted'),2)
f1score_nb = round(f1_score(y_test, y_pred_nb, average='weighted'),2)
specificity_nb = round(tn / (tn+fp), 2)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dt_classifier.fit(X_train, y_train)

y_pred_dt = dt_classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_dt).ravel()

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
akurasi_dt = round(accuracy_score(y_test, y_pred_dt),2)
precision_dt = round(precision_score(y_test, y_pred_dt, average='weighted'),2)
recall_dt = round(recall_score(y_test, y_pred_dt, average='weighted'),2)
f1score_dt = round(f1_score(y_test, y_pred_dt, average='weighted'),2)
specificity_dt = round(tn / (tn+fp), 2)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg_classifier = LogisticRegression(random_state = 0)
logreg_classifier.fit(X_train, y_train)

y_pred_logreg = logreg_classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_logreg).ravel()

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
akurasi_logreg = round(accuracy_score(y_test, y_pred_logreg),2)
precision_logreg = round(precision_score(y_test, y_pred_logreg, average='weighted'),2)
recall_logreg = round(recall_score(y_test, y_pred_logreg, average='weighted'),2)
f1score_logreg = round(f1_score(y_test, y_pred_logreg, average='weighted'),2)
specificity_logreg = round(tn / (tn+fp), 2)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state = 0)
rf_classifier.fit(X_train, y_train)

y_pred_rf = rf_classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_rf).ravel()

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
akurasi_rf = round(accuracy_score(y_test, y_pred_rf),2)
precision_rf = round(precision_score(y_test, y_pred_rf, average='weighted'),2)
recall_rf = round(recall_score(y_test, y_pred_rf, average='weighted'),2)
f1score_rf = round(f1_score(y_test, y_pred_rf, average='weighted'),2)
specificity_rf = round(tn / (tn+fp), 2)
```

### K-Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors = 7, metric = 'euclidean')
knn_classifier.fit(X_train, y_train)

y_pred_knn = knn_classifier.predict(X_test)

from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_knn).ravel()

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
akurasi_knn = round(accuracy_score(y_test, y_pred_knn),2)
precision_knn = round(precision_score(y_test, y_pred_knn, average='weighted'),2)
recall_knn = round(recall_score(y_test, y_pred_knn, average='weighted'),2)
f1score_knn = round(f1_score(y_test, y_pred_knn, average='weighted'),2)
specificity_knn = round(tn / (tn+fp), 2)
```

# **Model Evaluation**

```python
data_hasil_performa = pd.DataFrame({'Model':['Naive Bayes', 'Decision Tree', 'Logistic Regression', 'Random Forest', 'K-Nearest Neighbour'],
                                    'Akurasi':[akurasi_nb, akurasi_dt, akurasi_logreg, akurasi_rf, akurasi_knn],
                                    'Sensitivitas / Recall':[recall_nb, recall_dt, recall_logreg, recall_rf, recall_knn],
                                    'Spesifisitas':[specificity_nb, specificity_dt, specificity_logreg, specificity_rf, specificity_knn],
                                    'Precision':[precision_nb, precision_dt, precision_logreg, precision_rf, precision_knn],
                                    'F1-Score':[f1score_nb, f1score_dt, f1score_logreg, f1score_rf, f1score_knn]})

data_hasil_performa = data_hasil_performa.sort_values(by=['Akurasi', 'Precision', 'Sensitivitas / Recall', 'F1-Score', 'Model'], 
                                ascending=[False, False, False, False,True]).reset_index().drop(columns = 'index')
data_hasil_performa
```

```python
data_hasil_performa.to_excel('data_hasil_modelling.xlsx')
```
